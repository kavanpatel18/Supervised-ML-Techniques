```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor

# Training data
X = np.array([[0.2],
              [0.3],
              [1.3],
              [1.5],
              [2.0],
              [2.2]])

y = np.array([1.4, 1.3, 4.5, 4.7, 6.3, 6.5])

# KNN Regressor
model = KNeighborsRegressor(n_neighbors=3)

model.fit(X, y)

# New point
new_point = np.array([[1.4]])

prediction = model.predict(new_point)

print("Predicted Petal Length:", prediction[0])
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor

# Features: petal width, sepal length, sepal width
X = np.array([
    [0.2,5.1,3.5],
    [1.5,6.0,2.9],
    [1.8,6.5,3.0],
    [1.3,5.9,3.2]
])

# Target: petal length
y = np.array([1.4,4.5,5.5,4.2])

model = KNeighborsRegressor(n_neighbors=3)
model.fit(X,y)

new_point = np.array([[1.4,6.1,3.0]])

prediction = model.predict(new_point)

print("Predicted Petal Length:",prediction[0])
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# features: petal width, sepal length, petal length, sepal width
X = np.array([
[0.2,5.1,1.4,3.5],
[1.5,6.0,4.5,2.9],
[1.8,6.5,5.5,3.0],
[1.3,5.9,4.2,3.2]
])

y = ['Setosa','Versicolor','Virginica','Versicolor']

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

new_point = np.array([[1.4,6.1,4.4,3.0]])

print(model.predict(new_point))
```
