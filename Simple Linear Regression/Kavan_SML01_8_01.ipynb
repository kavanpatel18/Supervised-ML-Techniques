```python
#Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Loading Dataset
df=sns.load_dataset('iris')
df
```

```python
sns.set(color_codes=True)
```

```python
import warnings as w
w.filterwarnings("ignore")
```

```python
#Scatterplot
df.plot(kind='scatter', x='sepal_length', y='sepal_width')
```

```python
#Jointplot->Histograph+Scatterplot
sns.jointplot(x='sepal_length', y='sepal_width', data=df, kind='scatter')
```

```python
#Facetgrid
sns.FacetGrid(df, hue='species', palette='husl',height=5).map(plt.scatter, 'sepal_length','sepal_width').add_legend()
```

```python
sns.boxplot(x='species',y='sepal_length',palette='husl',data=df)
```

```python
a = np.arange(0,150)
df['ID']=a
df.columns
```

```python
sns.pairplot(df.drop('ID',axis=1), hue='species', palette='husl', size=4)
```
