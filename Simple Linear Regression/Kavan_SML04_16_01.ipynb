```python
#Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf('/content/SALES.txt', colspecs='infer', header=None) #Importing dataset
df
```

```python
#EDA
print(df.shape)
```

```python
df.head() #Top 4 rows
```

```python
df.columns = ['Advertising','Sales'] #renaming columns of df
df.head()
```

```python

df.info()
```

```python
df.describe()
```

```python
#Declare feature and target variable
x = df['Advertising'].values
y = df['Sales'].values
```

```python
#Visual EDA
plt.scatter(x,y, label='Scatter Plot')
plt.title("Relationship bw Sales and advertising")
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.legend(loc=4)
plt.show()
```

```python
print(x.shape)
print(y.shape)
#Dimensions of x and y
```

```python
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
#Reshaping x and y
print(x.shape)
print(y.shape) #New dimensions
```

```python
#Split x and y into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#Dimensions of x_train, x_test, y_train, y_test
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Fit the linear model
from sklearn.linear_model import LinearRegression
lm = LinearRegression() #Instantiate the linear regression object

lm.fit(x_train, y_train) #Train the model using training datasets

y_pred = lm.predict(x_test) #Make predictions using testing set
```

```python
a = lm.coef_
b = lm.intercept_
print("Estimated model slope: ",a)
print("Estimated model intercept: ",b)
#
```

```python
lm.predict(x) [0:5] #Predicting advertising values on first 5 sales values
```

```python
lm.predict(x)
```

```python
print(str(lm.predict(x)[24])) #Individual prediction
```

```python
#Calculate RMSE
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))
```

```python
#Calculate R2 score
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print("R2 score value: {:.4f}".format(r2))
```

```python
#Plot the regeression line
plt.scatter(x,y, label='Scatter Plot')
plt.plot(x_test, y_pred, color='black', linewidth=3, label = 'Regression line')
plt.title("Relationship bw Sales and advertising")
plt.xlabel('Advertising')
plt.ylabel('Sales')
plt.legend(loc=4)
plt.show()
```

```python
print("Training set score: {:.4f}".format(lm.score(x_train, y_train)))
print("Test set score: {:.4f}".format(lm.score(x_test, y_test)))
```
